```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
df = yf.download(['SPY', '^IXIC', '^DJI', '^RUA'], 
                 interval='1mo', start='1990-01-01', end='2024-01-01')[['Adj Close']]['Adj Close']
df = df.dropna()
df['SPY'] = np.log(df['SPY']).diff()*100
df['^RUA'] = np.log(df['^RUA']).diff()*100
df['^DJI'] = np.log(df['^DJI']).diff()*100
df['^IXIC'] = np.log(df['^IXIC']).diff()*100
# df['^TNX'] = df['^TNX'].diff()*100
# df['^TNX'] = np.log(df['^TNX'])*100
# df['^VIX'] = np.log(df['^VIX'])
df = df.dropna()
df
```

```python
df = df.rename(columns = {'SPY':'SP500', '^DJI':'Dow Jones', '^IXIC':'NASDAQ', '^RUA':'Russell 3000'})
df
```

```python
print(round(df.describe(),2).to_latex(caption='Table Title'))
```

```python
df = yf.download(['^TNX', '^FVX', '^TYX', '^VIX', 'SPY'], 
                 interval='1mo', start='1990-01-01', end='2024-01-01')[['Adj Close']]['Adj Close']
df = df.dropna()
df['^TNX'] = df['^TNX'].diff()*100
df['^FVX'] = df['^FVX'].diff()*100
df['^TYX'] = df['^TYX'].diff()*100
df['^VIX'] = np.log(df['^VIX'])
df = df.dropna()
df = df[['^TNX', '^FVX', '^TYX', '^VIX']]
df
```

```python
df = df.rename(columns = {'^TNX':'10Y USD', '^TYX':'30Y USD', '^FVX':'5Y USD', '^VIX':'VIX'})
df
```

```python
print(round(df.describe(),2).to_latex(caption='Table Title'))
```
